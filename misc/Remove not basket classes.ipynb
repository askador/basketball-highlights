```python
def remove_ext(filename):
  return filename.rsplit('.', 1)[0]
```

```python
def get_file_name(file_path):
  return file_path.rsplit('/', 1)[-1]
```

```python
def filter_labels(parent_path, split_name, basket_class_id):
  folder = os.path.join(parent_path, "labels")
  to_delete_file_names = []

  for filename in os.listdir(folder):
    label_path = os.path.join(folder, filename)
    with open(label_path, "r") as l:
      has_class = False
      lines = l.readlines()
      for line in lines:
        id = line.split(" ", 1)[0]
        has_class = int(id) == basket_class_id
    if not has_class:
      image_name = remove_ext(filename)
      to_delete_file_names.append(image_name)
      os.remove(label_path)
  return to_delete_file_names
```

```python
import shutil

def filter_basket_images(project, basket_class_id):
  src_path = os.path.join("/content", project)
  dest_path = os.path.join("/content", f"{project}_only-basket")
  shutil.copytree(src_path, dest_path, dirs_exist_ok=True)

  split_names = ["train", "test", "valid"]

  for split_name in split_names:
    split_dir = os.path.join(dest_path, split_name)
    if not os.path.isdir(split_dir):
      print(split_dir, "does not exist")
      continue

    to_delete_img_names = filter_labels(split_dir, split_name, basket_class_id)

    images_path = os.path.join(split_dir, "images")

    for filename in os.listdir(images_path):
      image_name = remove_ext(filename)
      if image_name in to_delete_img_names:
        image_path = os.path.join(images_path, filename)
        os.remove(image_path)
```

```python
filter_basket_images("Basketball-detection-1", 1)
```

```python
filter_basket_images("roi-dishith-device-faiz-sagar-2", 1)
```
